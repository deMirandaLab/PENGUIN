# Percentile filtering

```python
from scipy.ndimage import percentile_filter
```

```python
PERC = 50
grey_bool = np.where(grey_th > 0, 1, 0)
med = percentile_filter(grey_bool,percentile=PERC,footprint=np.ones((3, 3)))
bl = med.astype(bool) # False is Zero
grey_p = np.where(bl == False, 0, grey_th)

print(np.count_nonzero(grey_bool))
print(np.count_nonzero(grey_p))

# Display the combined array as an image with darker pixels closer to 1
plt.imshow(grey_p, cmap='gray_r')  # 'gray_r' is the reversed grayscale colormap
plt.colorbar()
plt.show()
```
